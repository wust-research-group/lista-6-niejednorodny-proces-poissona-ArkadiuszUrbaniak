```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def odwr_dyst(T=100):
    t, I = 0, 0
    S = [0]
    while t <= T:
        U = np.random.uniform(0, 1)
        t = (t + U) / (1 - U)
        if t <= T:
            I += 1
            S.append(t)

    n = len(S)
    N = np.arange(0, n)
    return S, N


S, N = odwr_dyst()

plt.scatter(S, N)
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()
```
