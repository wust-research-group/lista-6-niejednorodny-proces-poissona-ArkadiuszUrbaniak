```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def rozrzedzanie1(lambd=5, T=100):
    S = []
    I = 0
    t = 0
    while t <= T:
        U = np.random.uniform(0, 1)
        t = t - 1 / lambd * np.log(U)
        if t <= T:
            U2 = np.random.uniform(0, 1)
            if U2 <= (1 / (t + 1)) / lambd:
                I = I + 1
                S.append(t)
    n = len(S)
    N = np.arange(1, n + 1)
    return S, N


def rozrzedzanie2(lambd=5, T=100):
    S = []
    I = 0
    t = 0
    while t <= T:
        U = np.random.uniform(0, 1)
        t = t - 1 / lambd * np.log(U)
        if t <= T:
            U2 = np.random.uniform(0, 1)
            if U2 <= (2 / (t + 2)) / lambd:
                I = I + 1
                S.append(t)
    n = len(S)
    N = np.arange(1, n + 1)
    return S, N


S1, N1 = rozrzedzanie1()
S2, N2 = rozrzedzanie2()

plt.scatter(S2, N2)
plt.scatter(S1, N1, color="red")
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()

suma = np.sort(np.concatenate((S1, S2)))
n = len(suma)
N = np.arange(1, n + 1)

plt.scatter(suma, N)
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()
```
