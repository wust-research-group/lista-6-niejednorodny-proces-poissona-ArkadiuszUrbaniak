```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def rozrzedzanie(lambd=5, T=100):
    S = [0]
    I = 0
    t = 0
    while t <= T:
        U = np.random.uniform(0, 1)
        t = t - 1 / lambd * np.log(U)
        if t <= T:
            U2 = np.random.uniform(0, 1)
            if U2 <= (1 / (t + 1)) / lambd:
                I = I + 1
                S.append(t)
    n = len(S)
    N = np.arange(0, n)
    return S, N


S, N = rozrzedzanie()

plt.scatter(S, N)
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()
```
